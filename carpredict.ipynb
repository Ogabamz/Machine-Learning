```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("carprices.csv")
```

```python
df=df.rename(columns={"Car Model": "car_model",
                      "Sell Price($)":"price",
                      "Age(yrs)":"age"})
df
```

```python
dummies= pd.get_dummies((df.car_model), dtype=int)
dummies
```

```python
merged = pd.concat([df,dummies], axis = "columns")
merged
```

```python
final= merged.drop(["car_model"], axis ="columns")
final
```

```python
final= final.drop(['Mercedez Benz C class'], axis= 'columns')
```

```python
final
```

```python
X= final.drop('price', axis='columns')
X
```

```python
y= final.price
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
mercedes=model.predict([[45000,4,0,0]])
bmw=model.predict([[86000,7,0,1]])
print(mercedes)
print(bmw)
```
